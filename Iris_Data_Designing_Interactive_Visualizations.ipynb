## **1. Introduction to Interactive Visualizations**

---

### Why Interactivity?

Interactivity in data visualization transforms static charts and graphs into tools for exploration, making it easier to discover trends, outliers, and patterns within data. By adding interactive elements, viewers can dive deeper into the data and gain more personalized insights.

### Benefits of Interactivity:
- **Enables Exploration**: Interactive visualizations allow users to manipulate data and view different aspects of it, enhancing understanding and engagement. For example, a static chart might show general trends, but interactive filters or zoom features can help viewers hone in on specific areas of interest.
  
- **Increases User Engagement**: Interaction creates a dynamic experience, engaging users by allowing them to focus on specific data points or categories. Users can click, hover, or zoom to reveal additional information, which helps to hold their attention longer.

- **Provides Customized Insights**: By allowing users to adjust the view of data to their needs, interactivity supports more personalized and relevant data interpretation. For example, in a large dataset, users can use filters to isolate data for a specific time period, category, or range, tailoring the visualization to their objectives.

---

## Types of Interactivity

Interactive elements in visualizations typically fall into three main categories: filters, hover effects, and zoom features. Each of these types serves a distinct purpose in enhancing the viewer's experience.

### 1. Filters
- **Purpose**: Filters allow users to adjust the data displayed by selecting specific categories or ranges, thus focusing on relevant portions of the dataset.
- **Use Cases**: Commonly used in dashboards and exploratory data analysis tools, filters are ideal for breaking down large datasets by criteria such as date range, region, or demographic.
- **Example Applications**:
  - Filtering sales data by product category, region, or sales period.
  - Displaying weather data for selected cities or specific date ranges.

### 2. Hover Effects
- **Purpose**: Hover effects display additional details when a user hovers their cursor over an element, adding depth without cluttering the visualization.
- **Use Cases**: Ideal for providing contextual information or detailed breakdowns, hover effects are widely used in scatter plots, maps, and bar charts to reveal insights without overwhelming the primary view.
- **Example Applications**:
  - Showing detailed values or percentages on hover in a bar chart.
  - Displaying geographic information like city names or data points on a map when hovered over.

### 3. Zoom Features
- **Purpose**: Zoom functionality enables users to examine specific parts of the visualization in greater detail, especially useful for dense or time-series data.
- **Use Cases**: Zoom is particularly useful in line charts, maps, or scatter plots where examining data granularity is necessary.
- **Example Applications**:
  - Zooming into time-series data to view specific trends or peaks.
  - Using zoom on geographic maps to view city-level data within a country-level map.

# **Interactive Scatter Plot with Dropdown Menu Filter in Plotly**

In this example, we'll create an interactive scatter plot using Plotly, with a dropdown menu to filter data by species.

---

### Step 1: Import Libraries

First, we import `plotly.express` as `px`, a high-level interface for creating interactive plots with Plotly.

```python
import plotly.express as px
```

### Step 2: Load Sample Data

Plotly provides several built-in datasets. Here, we load the `iris` dataset, which contains measurements of iris flower species.
- **Features**: This dataset includes sepal width, sepal length, petal width, and petal length, along with species labels for each flower sample.



```python
# Load sample data
df = px.data.iris()
df.head()  # Display the first few rows of the dataset
```

```python
# prompt: Using dataframe df: matplotlib

import matplotlib.pyplot as plt

# Create the scatter plot
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.scatter(df['sepal_length'], df['sepal_width'], c=df['species_id']) # Color points by species
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs. Sepal Width')
plt.colorbar(label='Species ID') # Add a colorbar to indicate species ID mapping
plt.show()
```

### Step 3: Create a Basic Scatter Plot

We start by creating a scatter plot using `px.scatter`, setting:
- **x-axis** to `'sepal_width'`
- **y-axis** to `'sepal_length'`
- **color** to `'species'` (which will color the points based on the species type)

The resulting plot will show each data point as a dot, with colors representing different species.


```python
import matplotlib.pyplot as plt

# Create a basic scatter plot with species color differentiation
plt = px.scatter(df, x='sepal_width', y='sepal_length', color='species',
                 color_discrete_sequence=['#FF6347', '#4682B4', '#000000'])
plt.show()
```

### Step 4: Add a Dropdown Menu to Filter by Species

To make the plot interactive, we’ll add a dropdown menu. Each menu option will filter the plot by one of the three species or show all species at once.

#### Explanation of `update_layout`:
- **updatemenus**: This property adds interactive menu buttons to the plot.
- **buttons**: Each button in the dropdown menu is defined by:
  - **label**: The name displayed in the dropdown.
  - **method**: We use `"update"` to control visibility of data traces.
  - **args**: This specifies the visibility of each trace. Setting `"visible": [True, False, False]` shows only the first species, for example.

#### Button Configurations:
1. **Setosa**: Shows only `Setosa` species points.
2. **Versicolor**: Shows only `Versicolor` species points.
3. **Virginica**: Shows only `Virginica` species points.
4. **All**: Shows all species at once.

---



```python
# Add a dropdown filter to the scatter plot
plt.update_layout(
    updatemenus=[
        dict(
            buttons=[
                dict(label="Setosa", method="update", args=[{"visible": [True, False, False]}]),
                dict(label="Versicolor", method="update", args=[{"visible": [False, True, False]}]),
                dict(label="Virginica", method="update", args=[{"visible": [False, False, True]}]),
                dict(label="All", method="update", args=[{"visible": [True, True, True]}])
            ],
            direction="down"
        )
    ]
)
plt.show()
```

### 3.2 Slider Filter in Plotly

Use a slider to filter data based on continuous variables like date ranges or values.
Add a Range Slider to the X-Axis

Adding a range slider on the x-axis gives users an interactive tool to zoom in on specific ranges of the data.

#### Explanation:
- **fig.update_xaxes(rangeslider_visible=True)**: This command adds a range slider below the x-axis, allowing users to select a specific range within the plot. When users move the slider handles, the plot will automatically update to show only the data within the selected range.

Adding this functionality can make visualizations with continuous or time-based data easier to analyze and more dynamic.

---

```python
plt = px.scatter(df, x='sepal_width', y='sepal_length', color='species')
plt.update_xaxes(rangeslider_visible=True)
plt.show()
```

### Step 4: Enable Interactive Zooming and Panning on Both Axes

While Plotly doesn't have a direct y-axis slider, we can allow users to zoom and pan manually on both axes by enabling interactive controls.

### Explanation:
- **fig.update_layout(dragmode="pan")**: Enables panning across the plot.
- **range mode for y-axis**: We can specify initial ranges on the y-axis and allow users to zoom as needed.

### Additional Instructions:
To zoom, use the toolbar buttons in the top right of the plot:
- **Zoom**: Click and drag to zoom into a specific area.
- **Pan**: Move the plot without changing zoom level.

---


```python
plt = px.scatter(df, x='sepal_width', y='sepal_length', color='species')
plt.update_layout(dragmode="select")  # Allows dragging to pan across the plot
plt.update_yaxes(scaleanchor="x", scaleratio=0.1)  # Optional: lock aspect ratio if needed
plt.show()
```

# Interactive Drag Modes in Plotly

In Plotly, the `dragmode` parameter in `update_layout()` lets you control the type of interactive drag behavior in a plot. Here are the available drag modes:

- **`"zoom"` (default)**: Enables zooming by drawing a box over the area to zoom into.
- **`"pan"`**: Allows panning by clicking and dragging to move around the plot.
- **`"select"`**: Enables box selection, where you can click and drag a box to select multiple data points.
- **`"lasso"`**: Enables lasso selection, allowing you to freehand-select points in an arbitrary shape.
- **`"drawclosedpath"`**: Lets users draw a custom closed path (not typically used for data selection, but useful in annotation).
- **`"drawopenpath"`**: Enables drawing of an open path.
- **`"drawline"`**: Allows drawing straight lines on the plot.
- **`"drawrect"`**: Lets users draw rectangles on the plot.
- **`"drawcircle"`**: Enables circle drawing.
- **`"orbit"`**: Used for 3D plots, this mode allows users to rotate (orbit) the view around a 3D plot.
- **`"turntable"`**: Also for 3D plots, this mode enables rotation around a vertical axis, like a turntable.

# **Customizing Hover Text with `hovertemplate` in Plotly**
### 4.2 Customizing Hover Effects in Plotly

Customize hover information for a more detailed display.


In this example, we’ll customize the hover text for our scatter plot using `hovertemplate`. This allows us to control exactly what information appears in the tooltip when hovering over each data point.

---

### Step 1: Import Libraries and Load Data

We start by importing `plotly.express` as `px` and loading the sample `iris` dataset.


```python
import plotly.express as px

# Load the iris dataset
df = px.data.iris()
df.head()  # Display the first few rows of the dataset
```

### Step 2: Create a Basic Scatter Plot

We’ll create a scatter plot using:
- **x-axis** set to `'sepal_width'`
- **y-axis** set to `'sepal_length'`
- **color** set to `'species'` for differentiating species by color.

This initial plot has default hover text, displaying the x and y values and the color legend information.


```python
# Create a basic scatter plot
fig = px.scatter(df, x='sepal_width', y='sepal_length', color='species')
fig.show()
```

### Step 3: Customize Hover Text with `hovertemplate`

We use `update_traces` with the `hovertemplate` argument to create a customized tooltip format.

#### Explanation:
- **`fig.update_traces(hovertemplate="...")`**: This command allows us to specify the exact text to appear in each hover tooltip.
- **Syntax**:
  - `%{text}`: Represents custom text or additional data (such as the species name).
  - `%{x}` and `%{y}`: Refer to the x and y values (sepal width and sepal length, respectively) for each data point.

### Custom Hover Template:
In this example, we set `hovertemplate` to:
- `"Species: %{text}<br>Sepal Width: %{x}<br>Sepal Length: %{y}"`

#### Tooltip Display:
1. **Species**: Shows the species name (`%{text}` placeholder).
2. **Sepal Width**: Displays the value for the x-axis (`%{x}`).
3. **Sepal Length**: Displays the value for the y-axis (`%{y}`).
4. **Line Breaks**: `<br>` is used to create line breaks for a clean, organized layout.

---



```python
# Customize hover text for each data point
fig.update_traces(hovertemplate="Species: %{species}<br>Sepal Width: %{x}<br>Sepal Length: %{y}")
fig.show()
```

# **Creating an Interactive Plot with Hover Tool in Bokeh**
### 4.3 Hover Effects in Bokeh

We can also use Bokeh’s `HoverTool` for detailed hover effects.


In this example, we will use Bokeh to create a scatter plot with a hover tool, displaying additional information when hovering over each data point.

---

### Step 1: Import Libraries and Set Up Output Display

1. **Bokeh Library Imports**:
   - `figure` and `show` from `bokeh.plotting` to create and display the plot.
   - `output_notebook` to ensure that the plot displays inline in a Jupyter Notebook.
   - `HoverTool` from `bokeh.models` to add interactive hover functionality.

2. **output_notebook()**: This function configures Bokeh to display plots within the notebook rather than in a separate HTML file.

---



```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool

# Configure output to display plots in the notebook
output_notebook()
```

### Step 2: Create a Basic Scatter Plot

We create a plot with the following properties:
- **figure(title="...")**: Initializes the plot with a title ("Hover Tool Example").
- **p.circle**: Draws circular markers on the plot.
  - `[1, 2, 3]` for x-coordinates.
  - `[4, 5, 6]` for y-coordinates.
  - `size=15`: Sets circle size to 15.
  - `color="navy"`: Colors the circles navy blue.
  - `alpha=0.5`: Sets transparency to 50% for a softer look.



```python
# Create the figure and add circle glyphs
p = figure(title="Hover Tool Example")
p.scatter([1, 2, 3], [4, 5, 6], size=15, color="navy", alpha=0.5)
```

### Step 3: Add the Hover Tool

1. **hover = HoverTool()**: Creates a HoverTool instance.
2. **hover.tooltips**: Defines the tooltip content.
   - **Syntax**: `("label", "@field")` pairs:
     - `"X-Value"`: Label displayed in the tooltip for the x-coordinate.
     - `@x` and `@y`: These are special fields in Bokeh representing the x and y coordinates of each point.

3. **p.add_tools(hover)**: Adds the hover tool to the plot, enabling interactive tooltips.

When you hover over a data point, the tooltip will display the x and y values.

---



```python
# Set up the HoverTool and add it to the plot
hover = HoverTool()
hover.tooltips = [("X-Value", "@x"), ("Y-Value", "@y")]
p.add_tools(hover)
```

### Step 4: Display the Plot

Finally, we use `show(p)` to render the plot with the hover tool enabled.



```python
# Display the plot
show(p)
```

# **Adding a Dropdown Widget to a Bokeh Plot**

## 6. Advanced Custom Interactivity with Bokeh Widgets

Widgets in Bokeh allow us to add dropdowns and sliders for even more interactivity.

### Dropdown and Slider Example


In this example, we will add a dropdown widget (`Select`) to a Bokeh plot. This allows users to select a species from a dropdown menu, and interactively filter or update the plot based on their selection.

---

### Step 1: Import Libraries and Set Up the Plot

We start by importing the necessary libraries:
- **`figure`, `show`** from `bokeh.plotting` to create and display the plot.
- **`column`** from `bokeh.layouts` to arrange the widgets and the plot vertically.
- **`Select`** from `bokeh.models` to create the dropdown widget.
- **`Slider`** is imported in case you'd like to add a slider in the future for interactive adjustments (not used directly in this example, but available for extension).

We also call `output_notebook()` to ensure the plot displays inline in a Jupyter notebook.



```python
pip install bokeh_sampledata
```

```python
# Importing required libraries
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, Select
from bokeh.layouts import column
from bokeh.sampledata.iris import flowers

output_notebook()
```

### Step 2: Create the Basic Plot

We create a simple scatter plot (`p`) to visualize data. For this example, we assume that the plot is showing some data points, and we are using the iris dataset (as in previous examples).



```python
# Prepare the data
df = flowers.copy()
df['species'] = df['species'].astype(str)

# ColumnDataSource for filtering
source = ColumnDataSource(data=df)

# Create the scatter plot
p = figure(title="Scatter Plot of Iris Data", width=700, height=500,
           x_axis_label='Sepal Width', y_axis_label='Sepal Length')

scatter = p.scatter(x='sepal_width', y='sepal_length', source=source, size=10, color='navy', alpha=0.6)
```

### Step 3: Create the Dropdown Widget

1. **`Select()`**: The `Select` widget creates a dropdown menu. In this case, the dropdown allows users to choose between three species of iris flowers: "setosa," "versicolor," and "virginica."
   
   - **`title="Select Species"`**: This sets the label of the dropdown.
   - **`value="setosa"`**: This is the default value when the dropdown is first rendered (set to "setosa").
   - **`options=["setosa", "versicolor", "virginica"]`**: These are the available options for the user to select from.

The dropdown widget will allow users to select one of the species from the list.



```python
# Dropdown menu for filtering by species
dropdown = Select(title="Select Species:", value="All", options=["All"] + df['species'].unique().tolist())

# Callback function to update the data
def update_plot(attr, old, new):
    if dropdown.value == "All":
        source.data = df
    else:
        source.data = df[df['species'] == dropdown.value]

# Link the dropdown menu to the callback
dropdown.on_change('value', update_plot)


```

### Step 4: Arrange and Display the Plot and Widget

We use **`column()`** to arrange the dropdown widget and the plot vertically, so that the widget appears above the plot.

- **`column(dropdown, p)`**: This combines the dropdown and the plot in a vertical layout.

Finally, we use **`show()`** to display the plot and the dropdown together.



```python
# Arrange the dropdown and plot in a vertical layout
show(column(dropdown, p))
```

```python
# Install Bokeh if not already installed
!pip install bokeh --quiet

# Importing required libraries
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, Select, CustomJS
from bokeh.layouts import column
from bokeh.sampledata.iris import flowers

output_notebook()

# Prepare the data
df = flowers.copy()
df['species'] = df['species'].astype(str)

# Create ColumnDataSource for all data and filtered data
source = ColumnDataSource(data=df)

# Create the scatter plot
p = figure(title="Scatter Plot of Iris Data", width=700, height=500,
           x_axis_label='Sepal Width', y_axis_label='Sepal Length')

p.scatter(x='sepal_width', y='sepal_length', source=source, size=10, color='navy', alpha=0.6)

# Dropdown menu for filtering by species
dropdown = Select(title="Select Species:", value="All", options=["All"] + df['species'].unique().tolist())

# JavaScript callback for updating the plot
callback = CustomJS(args=dict(source=source, full_data=df), code="""
    const selected_species = cb_obj.value;  // Get the selected species
    const data = full_data;

    if (selected_species === "All") {
        source.data = data;  // Show all data
    } else {
        const filtered = {
            sepal_width: [],
            sepal_length: [],
            species: []
        };

        // Filter data based on the selected species
        for (let i = 0; i < data['species'].length; i++) {
            if (data['species'][i] === selected_species) {
                filtered['sepal_width'].push(data['sepal_width'][i]);
                filtered['sepal_length'].push(data['sepal_length'][i]);
                filtered['species'].push(data['species'][i]);
            }
        }

        source.data = filtered;  // Update the source with filtered data
    }
    source.change.emit();  // Trigger update
""")

dropdown.js_on_change('value', callback)  # Link dropdown to the callback

# Layout with dropdown and plot
layout = column(dropdown, p)

# Show the interactive visualization
show(layout)
```
